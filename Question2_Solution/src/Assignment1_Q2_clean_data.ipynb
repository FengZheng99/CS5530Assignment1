```python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the CSV file
df = pd.read_csv('/content/StudentsPerformance.csv')
df.head()
```

###**Data Preprocessing**

```python
# find unique value each column
for col in df:
  print(df[col].unique())
```

```python
# find total number of null value
df.isnull().sum()
```

```python
# reduce spaces in cloumn name by rename it
df = df.rename(columns={'parental level of education':'parental_education', 'test preparation course':'preparation_course',
                   'math score':'math_score', 'reading score':'reading_score', 'writing score':'writing_score'})
df
```

```python
# create an average score column
df['average_score'] = df[['math_score', 'reading_score', 'writing_score']].mean(axis=1).round(2)
df.head()
```

```python
# write the cleaned csv file
df.to_csv('clean_StudentsPerformance.csv')
```
