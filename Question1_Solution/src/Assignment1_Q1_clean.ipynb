```python
# import necessary libraries
import pandas as pd
```

```python
# read the CSV file
df = pd.read_csv('/content/Assign1_Q1_data_raw.csv')
df.head()
```

```python
# find unique value each column
for col in df:
  print(df[col].unique())
```

###**Data Preprocessing**

```python
# find total number of null value
df.isnull().sum()
```

```python
# reduce spaces in cloumn name by rename it
df = df.rename(columns={'Height (Inches)':'Height_inches', 
                        'Weight (Pounds)':'Weight_pounds',
                        'Grip strength':'Grip_strength'})
df
```

```python
# create an BMI column
df['BMI'] = (df['Weight_pounds']/(df['Height_inches'])**2*703).round(2)
df.head()
```

```python
# write the cleaned csv file
df.to_csv('Assign1_Q1_data_clean.csv')
```
